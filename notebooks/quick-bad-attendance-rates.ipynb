# Bad attendance rates

Some of the values from our 2016 schools dataset are wrong (they are reporting 0% attendance rate, when it is actually unkown). We should convert these values to NaN.

```python
import pandas as pd

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('../data/raw/2016 School Explorer.csv')
df.head()
```

```python
a = df['Student Attendance Rate'] == '0%'
a.sum()
```

```python
b = df['Percent of Students Chronically Absent'] == '100%'
b.sum()
```

```python
(a == b).all()
```

Both columns are wrong on the same places.

```python
df[a]
```

As we can see, the bad values are from the "SUCCESS ACADEMY CHARTER SCHOOL"... Lots of them.

If we take a look at these values online:

https://tools.nycenet.edu/guide/2016/#dbn=84K367&report_type=EMS

The "Percentage of students with 90%+ attendance" is not 0%, but actually, it is not shown. Given that school is still working (all other scores are still there),
we are safe to assume that this value, instead of being 0%, is not known.
