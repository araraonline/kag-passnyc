### Columns

- SED Code (ID): State Education Department code
- District (ID): City district
- Latitude/Longitude: Coordinates
- Zip (ID): USA Postal code
- Average ELA Proficiency/Average Math Proficiency: seem like aggregate measures
- Economic Need Index:

  The Economic Need Index reflects the socioeconomics of the school
  population. It is calculated using the following formula:

  Economic Need Index = (Percent Temporary Housing) + (Percent
  HRA-eligible * 0.5) + (Percent Free Lunch Eligible * 0.5)

\*: ELA means English Language Arts

## District 5 SHSAT

It may be a valuable resource for PASSNYC by itself or may be used as a base to predict enrollment in another districts.

Data is clean without missing input.
