# Preparing

## Tests DataFrame

```python
# since there were only 5 students tested in Math, the results disappear
# this is a sign language school, it is safe to remove from the dataset later

tests = pd.read_pickle('../data/pre/test_results.pkl')  # NYS Test Results
tests[(tests['DBN'] == '02M347') &
      (tests['Grade'] == 7) &
      (tests['Year'] == 2017)].iloc[0]
```

```python
tests = pd.read_pickle('../data/pre/test_results.pkl')  # NYS Test Results
tests = tests.set_index('DBN')
tests = tests[(tests['Grade'] == 7) &
              (tests['Year'] == 2017)]
tests = tests[[
    'Charter School',
    'Mean Scale Score - ELA',
    '% Level 2 - ELA',
    '% Level 3 - ELA',
    '% Level 4 - ELA',
    'Mean Scale Score - Math',
    '% Level 2 - Math',
    '% Level 3 - Math',
    '% Level 4 - Math',
]]

tests.head()
```

```python
tests.shape
```

# Merging

```python
joined = shsat.join(base, how='inner').join(demographics, how='inner').join(tests, how='inner')  # left joins
joined.head()
```
