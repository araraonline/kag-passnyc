```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_pickle('../data/process/schools2017.pkl')
df = df.drop([
    '# SHSAT Offers',
    '# SHSAT Testers',
    '% SHSAT Offers',
    '% SHSAT Testers'
], axis=1)
```

```python
def nulltuple(row):
    return tuple(row[row.isnull()].index.tolist())

df.apply(nulltuple, axis=1).value_counts().to_frame()
```

```python
# these two schools are missing everything

# First one is pretty new, for this reason it has no data available
# Second one is New York Center for Autism Charter School

df.loc[df['School Name'].isnull()]
```

```python
# no ELA or Math information... It will be hard fitting a model

df.loc[
    df['School Name'].notnull() &
    df['Mean Scale Score - ELA'].isnull()
]
```

```python
# no Math information... We can rely on ELA

df.loc[
    df['School Name'].notnull() &
    df['Mean Scale Score - ELA'].notnull() &
    df['Mean Scale Score - Math'].isnull()
]
```

```python
# no presence information
# they will get almost a full-fledged model

df.loc[
    df['School Name'].notnull() &
    df['Mean Scale Score - ELA'].notnull() &
    df['Mean Scale Score - Math'].notnull() &
    df['Percent of Students Chronically Absent'].isnull()    
]

# success accademy charter school ...
```
