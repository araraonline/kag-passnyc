# (prototype) Model 1 - Results Visual DataFrame

In this notebook I will convert the results from model 1 into a beautiful dataframe.

```python
import pandas as pd
```

```python
def join(s1, s2):
    if s1 == s2:
        return s1
    else:
        return ' ðŸ ’ '.join([s1, s2])
    
f_0 = lambda f: '{:.0f}'.format(f)
f_2 = lambda f: '{:.2f}'.format(f)
f_pct = lambda f: '{:.0%}'.format(f)

cnt_columns = [
    'Actual # (min)',
    'Actual # (max)',
    'Estimated #',
    'Difference # (min)',
    'Difference # (max)',
]

pct_columns = [
    'Actual % (min)',
    'Actual % (max)',
    'Estimated %',
    'Difference % (min)',
    'Difference % (max)',
]

join_columns = [
    ('Actual # (min)', 'Actual # (max)'),
    ('Difference # (min)', 'Difference # (max)'),
    ('Actual % (min)', 'Actual % (max)'),
    ('Difference % (min)', 'Difference % (max)'),
]

df = pd.read_pickle('../data/output/model1_results.pkl')
df['Chance Multiplier'] = df['Chance Multiplier'].apply(f_2)
for c in cnt_columns:
    df[c] = df[c].apply(f_0)
for c in pct_columns:
    df[c] = df[c].apply(f_pct)
for c1, c2 in join_columns:
    name = c1[:-6]
    column = [join(x1, x2) for x1, x2 in zip(df[c1], df[c2])]
    df[name] = column
    df = df.drop([c1, c2], axis=1)
    
df = df[[
    'Actual #',
    'Estimated #',
    'Difference #',
    
    'Actual %',
    'Estimated %',
    'Difference %',
    
    'Chance Multiplier',
]]
df = df.applymap(lambda x: "{:>10}".format(x))

df.head()
```
