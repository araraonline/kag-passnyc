# (prototype) Join demographics data over the years

```python
import pandas as pd
```

```python
sd2015 = pd.read_pickle('../data/pre/school_demographics_20142015.pkl')
sd2016 = pd.read_pickle('../data/pre/school_demographics_20152016.pkl')
sd2017 = pd.read_pickle('../data/pre/school_demographics_20162017.pkl')
```

```python
sd2015['Year'] = 2015
sd2016['Year'] = 2016
sd2017['Year'] = 2017
```

```python
sd2015 = sd2015.reset_index().set_index(['DBN', 'Year'])
sd2016 = sd2016.reset_index().set_index(['DBN', 'Year'])
sd2017 = sd2017.reset_index().set_index(['DBN', 'Year'])
```

```python
sd = pd.concat([sd2015, sd2016, sd2017])
sd.head()
```
