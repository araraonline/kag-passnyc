# Robust Regression vs. Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import HuberRegressor, LinearRegression
```

```python
np.random.seed(100)

# generate sample dataset
xs = np.random.normal(size=100)
err = np.random.normal(size=100) / 3
ys = 5 + 1.2 * xs + err

# add outliers
xs = np.append(xs, [1.6, 1.65, 1.7, 1.8])
ys = np.append(ys, [11.4, 11.5, 10.9, 11.7])

plt.plot(xs, ys, '.')
```

```python
inputs = xs.reshape(-1, 1)  # entries must be on the rows
outputs = ys

model1 = LinearRegression()
model1.fit(inputs, outputs)
predictions1 = model1.predict(inputs)

model2 = HuberRegressor(epsilon=1)
model2.fit(inputs, outputs)
predictions2 = model2.predict(inputs)

plt.figure(figsize=(8,5))
plt.plot(xs, predictions1)
plt.plot(xs, ys, '.')
plt.title("Model #1")

plt.figure(figsize=(8,5))
plt.plot(xs, predictions2)
plt.plot(xs, ys, '.')
plt.title("Model #2")
```
